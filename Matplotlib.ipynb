```python
#we need the subpackage pyplot of matplotlib to visualize data
import matplotlib.pyplot as plt
```

```python
#lets get some insight in the evolution of the world population
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
```

```python
#basic line plot
#to plot this data as a line chart we call plt.plot
plt.plot(year, pop)
#the first argument corrosponds to horizontal axis and the second one to the vertical

#but the pot will not pop up until we call the show function to display the chart
plt.show()

#overall, plot function tells python what to plot and how to plot
#show actually displays the plot
```

```python
#Scatter plot
#there's one more plotting technique and that is scatter plot, to use this
plt.scatter(year, pop)
plt.show()
```

```python
#now if we want to do the same but put the x-axis on a loagarithmic scale, we use plt.scale('log')
plt.scatter(year, pop)
plt.xscale('log')
plt.show()
```

```python
#Histogram
#histogram is another data visualization method from the matplotlib package
#it helps to get an idea about the distribution of the variables

values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
#here hist takes two parameters one is the list of values and the other is the number of bins it is going to divide into
plt.hist(values, bins=3)
#if we do not set the number of bins, python will set the number of bins to 10 by default
plt.show()
```

```python
plt.hist(values, bins=3)
plt.show()
#here clf function is used to clear plot again
plt.clf()

plt.hist(values, bins=2)
plt.show()
plt.clf()
```

```python
#customization of the plot
#above plots are very basic and we can give more data to the reader if we customize the plot

#but first let's fill two arrays for the plot
new_year = [1951]
for i in range(1952, 2100):
    new_year.append(i)
    
#randomly values are inserted in new_pop with the help of numpy funcion
import numpy as np
new_pop = np.round(np.random.normal(1.752, 0.5, 149), 3)
new_pop.sort(); 

plt.plot(new_year, new_pop)

#if we want to use label for x and y axis we do this
plt.xlabel('Year')
plt.ylabel('Population')

#we can also add a title for the plot
plt.title("World Population Projections")

#now if i want the y-axis to start from zero upto four with itervals of two
#add to that is a list that is a display names of the ticks 
plt.yticks([0, 2, 4], ['0', '2B', '4B'])

plt.show()
```
